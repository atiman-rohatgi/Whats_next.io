```python
# Define a function to search for a string in a specific column
def search(dataframe, column_name, search_string):
    return dataframe[dataframe[column_name].str.contains(search_string, case=False, na=False)]

# Example usage
search(df_pc, "name", "grand")
```

```python
# Group by 'title' and 'release_date' and consolidate 'ps4' and 'ps5' columns
df_ps = df_ps.groupby(["title", "release_date"]).agg({
    "ps4": "max",  # If any row has True, keep True
    "ps5": "max",  # If any row has True, keep True
    "gameid": "first",  # Keep the first occurrence of gameid
    "publishers": "first",  # Keep the first occurrence of publishers
    "genres": "first"  # Keep the first occurrence of genres
}).reset_index()

```

```python
# Remove spaces from the ends of the 'title' column in df_ps
df_ps["title"] = df_ps["title"].str.strip()

# Remove spaces from the ends of the 'name' column in df_pc
df_pc["name"] = df_pc["name"].str.strip()
```

```python
df_pc.head(10)
```

```python
import re
```

```python
# Replace '-' with a space in 'title' in df_ps
df_ps["title"] = df_ps["title"].str.replace("-", " ", regex=False)

# Replace '-' with a space in 'name' in df_pc
df_pc["name"] = df_pc["name"].str.replace("-", " ", regex=False)

df_pc["name"] = df_pc["name"].str.replace(":", " ", regex=False)

df_ps["title"] = df_ps["title"].str.replace(r'[.,;"@#$%&*()!]', ' ', regex=True)

df_pc["name"] = df_pc["name"].str.replace(r'[.,;:"@#$%&*()!]', ' ', regex=True)
```
